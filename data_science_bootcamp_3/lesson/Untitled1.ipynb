```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(111, projection = '3d')
```

```python
bikes = pd.read_csv('data/bikeshare.csv', index_col='datetime', parse_dates = True)
```

```python
bikes = bikes.rename(columns={'count':'total_rentals'})
```

```python
feature_cols = ['temp','humidity']
```

```python
from sklearn.linear_model import LinearRegression

# Create X and y
X = bikes[feature_cols]
y = bikes['total_rentals']

# Instantiate your model 
lr = LinearRegression()

# fit your model
lr.fit(X, y)          

#print intercept and coeffients
print(lr.intercept_)
print(lr.coef_)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
z = bikes.total_rentals
x = bikes.temp
y = bikes.humidity
ax.scatter(x, y, z, marker='o')
ax.plot(x,y, lr.predict(X))
```
