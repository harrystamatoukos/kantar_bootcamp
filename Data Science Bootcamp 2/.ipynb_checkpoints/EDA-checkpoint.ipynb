# Section 2: Filtering and Sorting

Logical Filtering:

```python
users
```

```python
under_20 = users['age'] < 20 
```

```python
users[under_20]
```

```python
users[users['age'] < 20]
```

```python
mask = users.occupation == 'student'
```

```python
mask
```

```python
users[mask]
```

```python
#Important note 
users
```

```python
students = users[mask]
```

```python
students.head()
```

Logical Filtering with multiple conditions

```python
drinks
```

```python
# What are the european countries that drink a lot of beer
mask = (drinks.beer_servings > 25) & (drinks.continent =='EU')
europe = drinks[mask]
```

```python
drinks[mask]
```

```python
# OR
mask = (drinks.beer_servings > 25) | (drinks.continent =='EU')
drinks[mask]
```

```python
mask = ((drinks.beer_servings > 25) & (drinks.continent =='EU')) | (drinks.wine_servings > 200)
```

```python
mask
```

```python
users[users.occupation.isin(['student','doctor'])]
```
