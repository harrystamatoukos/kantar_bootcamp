# Split - Apply - Combine

- Split : we group data by continent
- Apply : for each group we would apply the mean() function
- Combine : combine the group and the mean() function

```python
drinks = pd.read_csv('data/drinks.csv')
```

```python
drinks
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
#For each continent to calculate the average for every column
drinks.groupby('continent').mean()
```

```python
drinks.groupby('continent').beer_servings.describe()
```

```python
# Agg allows us to aggregate results of a series function
drinks.groupby('continent').beer_servings.agg(['mean','count','max'])
```

```python
#We can sort 
drinks.groupby('continent').beer_servings.agg(['mean','count','max']).sort_values('mean', ascending = False)
```

```python
drinks.groupby('continent').continent.count()
```

```python
# Create a function that takes an inpute and rounds it to the nearest whole number
# round(x,0)
```

```python
def round_whole(x):
    '''
    INPUT: this functions takes input a sinlge number
    OUTPUT: the outpus is a rounded number
    '''
    rounder = round(x,0)
    return rounder
```

```python
drinks['liters'] = drinks.liters.apply(lambda x: int(round_whole(x)) )
```
